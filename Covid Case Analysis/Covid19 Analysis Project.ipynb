# Case Study : Covid Cases Data Analysis

```python
#Importing Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the Dataset
dataset= pd.read_csv("covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
#Parmanent Deleting the columns from the DataSet
df=dataset.drop(['Lat','Long'], axis=1, inplace=True)
dataset.head()
```

```python
#see countrywise aggregate data
corona_Agg_dataset=dataset.groupby('Country/Region').sum()
corona_Agg_dataset.head()
```

```python
# After Grouping
corona_Agg_dataset.shape
```

```python
#Visualise the Data
corona_Agg_dataset.loc["India"].plot()
corona_Agg_dataset.loc["Afghanistan"].plot()
corona_Agg_dataset.loc["US"].plot()
corona_Agg_dataset.loc["Japan"].plot()
plt.legend()
```

```python
#last 5 days corona spread data
corona_Agg_dataset.loc["India"][-5:].plot()
```

```python
corona_Agg_dataset.loc["India"].diff().plot()
```

```python
# Max case reported
corona_Agg_dataset.loc["India"].diff().max()
```

```python
# Max case reported
corona_Agg_dataset.loc["Spain"].diff().max()
```

```python
countrys=list(corona_Agg_dataset.index)

df=pd.DataFrame(corona_Agg_dataset)
df = df.apply(pd.to_numeric, errors='coerce')
max_inf_rate= []

for c in countrys:
    max_inf_rate.append(df.loc[c].diff().max())
corona_Agg_dataset["Max_infection_Rate"]=max_inf_rate
```

```python
corona_Agg_dataset
```

```python
corona_data=pd.DataFrame(corona_Agg_dataset["Max_infection_Rate"])
corona_data
```

```python
#importing Dataset 2 ie World Happiness Report
happines_report=pd.read_csv("worldwide_happiness_report.csv") 
```

```python
happines_report
```

```python
#drop useless column

useless=['Overall rank','Score','Generosity','Perceptions of corruption']
happines_report.drop(useless,axis=1, inplace=True)
happines_report
```

```python
happines_report.set_index('Country or region',inplace=True)
happines_report.head()
```

```python
happines_report.shape
```

```python
finalData=corona_data.join(happines_report, how="inner")
finalData
```
